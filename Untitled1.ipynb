```python
%matplotlib inline
import numpy as np
import torch
torch.set_printoptions(edgeitems=2, linewidth=75)
```

```python
t_c = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
```

```python
def linear_model(t_u, w, b):
    return w * t_u + b 
```

```python
def loss_fn(t_p, t_c):
    squared_diff = (t_p - t_c) ** 2
    return squared_diff.mean()
```

```python
w = torch.ones(())
b = torch.zeros(())
```

```python
t_p = linear_model(t_u, w, b)
t_p
```

```python
loss = loss_fn(t_p, t_c)
loss
```

### Broadcasting:
Following rules to match the tensors:
1. For each index dimension, counted from the back, if one of the operands is size 1 in theat dimension, PyTorch will use the single entry along this dimension with each of the entries in the other tensor along this dimension. 
2. If both sizes are greater than 1, they must be the same, and natural matching must be used
3. If one of the tensors has mroe index dimensions other than the other, the entirety of the other tensor will be used for each entry along these dimensions.

```python
x = torch.ones(())
y = torch.ones(3,1)
z = torch.ones(1,3)
a = torch.ones(2, 1, 1)
print(f"shapes: x: {x.shape}, y: {y.shape}")
print(f"        z: {z.shape}, a: {a.shape}")
print("x * y:", (x * y).shape)
print("y * z:", (y * z).shape)
print("y * z * a:", (y * z * a).shape)
```

```python
delta = 0.1
loss_rate_of_change_w = \
(loss_fn(linear_model(t_u, w + delta, b), t_c) - loss_fn(linear_model(t_u, w - delta, b), t_c)) / (2.0 * delta) 
```

```python
learning_rate = 1e-2
w = w - learning_rate * loss_rate_of_change_w
```

```python
loss_rate_of_change_b =  \
    (loss_fn(linear_model(t_u, w , b + delta), t_c) - 
     loss_fn(linear_model(t_u, w , b - delta), t_c)) / (2.0 * delta) 
```

```python
b = b - learning_rate * loss_rate_of_change_b
```

```python
def dloss_fn(t_p, t_c):
    dsq_diffs = 2 * (t_p - t_c) / t_p.size(0)  # <1> TP.size is for the derivative of the mean
    return dsq_diffs
```

```python
def dmodel_dw(t_u, w, b):
    return t_u
```

```python
def dmodel_db(t_u, w, b): 
    return 1 
```

#### Now we can define the gradient function:

```python
def grad_fn(t_u, t_c, t_p, w, b):
    dloss_dtp = dloss_fn(t_p, t_c)
    dloss_dw = dloss_dtp * dmodel_dw(t_u, w, b)
    dloss_db = dloss_dtp * dmodel_db(t_u, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])
```

### Full training Loop

Parameters needed:
1. epochs
2. learning rate
3. variables
4. parameters



Steps:
1. Initialise
2. Forward prop
4. Calculate loss
3. Backward prop
4. Repeat 

```python
def training_loop(n_epochs, learning_rate, params, t_u, t_c,print_params=True):
    for epoch in range(1, n_epochs + 1):
        w, b = params ## init
        
        t_p = linear_model(t_u, w, b) ## Forward Prop
        loss = loss_fn(t_p, t_c) ### Loss function
        grad = grad_fn(t_u, t_c, t_p, w, b) ## Backward pass
        
        params =  params - learning_rate * grad
        
        if epoch in {1, 2, 3, 10, 11, 99, 100, 4000, 5000}:  # <3>
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
            if print_params:
                print('    Params:', params)
                print('    Grad:  ', grad)
        if epoch in {4, 12, 101}:
            print('...')

        if not torch.isfinite(loss).all():
            break  # <3>
            

        
    return params


```

```python
training_loop(
    n_epochs = 100, 
    learning_rate = 1e-2, 
    params = torch.tensor([1.0, 0.0]), 
    t_u = t_u, 
    t_c = t_c)
```

##### Normalising Inputs
When?
1. When the gradients of the weights are very different

Why do we normalise inputs?
1. To increase the speed of 
2. If different gradiants, then the wieghts and bias live in differently scaled spaces. Therefore, a learning rate that is large enough to meaningfully update one will be so large as to be unstable for the other. The same rate most likely won't be applicable to both
3. Improves model convergence (see example above)

#### Computing the gradient automatically: (AutoGrad)
1. Pytorch tensors can remember where they come from, in terms of the operations and other parent tensors that originated them, an d they can automatically provide the chain of derivatives of such operations with respect to their inputs
2. The Grad Attribute: The arguement is telling pytorch to track the entire family tree of tensors resulting from operations on params. Anmy tensor that will have params as an ancestor will have acces to the chain of functions that were called to get from params to that tensor. 
3. If these functions are differentiatable then value of the derivative will be automaticall populated as a grad attibuted of the params tensor.

```python
params = torch.tensor([1.0, 0.0], requires_grad=True)
params.grad is None
```

```python
loss = loss_fn(linear_model(t_u, *params), t_c)
```
